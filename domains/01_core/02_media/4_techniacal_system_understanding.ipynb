#### Typical Pipeline

```tree
News API / CMS
   ↓
Article Published Events
   ↓
Kafka (stream ingestion)
   ↓
Spark / Streaming Engine
   ↓
Processing:
   - Trend detection
   - Sentiment analysis
   ↓
Storage (HDFS / Warehouse)
   ↓
Dashboard / Recommendation Engine

___

- Real-time Importance
  -  Medium to High
  - Trending topics need near real-time
  - Not as critical as finance
