### Data understanding


- In e-commerce, data is not static tables first (like Excel sheets).Instead, it is generated from user actions happening over time.
- understand
  - What actions exist?
  - What data gets generated from each action?
  - How these actions connect to business outcomes (revenue, retention, etc.)
  

___

#### **Step-by-step user journey:**
- User opens app
  - Event: app_open
- User searches “shoes”
  - Event: search
- User clicks a product
  - Event: product_click
- User adds to cart
  - Event: add_to_cart
- User purchases
  - Event: purchase

___

- Instead of a clean “table of customers”, you mostly get something like:

| user_id | event_name    | timestamp | metadata       |
| ------- | ------------- | --------- | -------------- |
| 101     | search        | 10:01     | query=shoes    |
| 101     | product_click | 10:02     | product_id=567 |
| 101     | add_to_cart   | 10:03     | product_id=567 |
| 101     | purchase      | 10:05     | amount=2000    |


- This is called event log data

___

#### **You must:**
- Reconstruct behavior
  - Build user journeys from events
- Derive metrics
  - Conversion rate = purchases / visits
  - Drop-off rate = users who didn’t continue
-  Create business tables
   -  Funnel table
   -  Cohort table
   -  Retention table
