### Pipeline

```tree
User Activity (App)
   ↓
Event Stream (Kafka)
   ↓
Real-time Processing (Spark/Flink)
   ↓
Use Cases:
   - Feed ranking
   - Trend detection
   - Spam detection
   ↓
Storage (HDFS / DB)
   ↓
Feed API / Dashboard

___

- Real-time Importance
  - VERY HIGH
  - Feed ranking must be instant
  - Notifications are real-time
