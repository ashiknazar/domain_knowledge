#### ML / AI Use Cases

___

-  Feed Ranking (MOST IMPORTANT)
   - Decide what user sees first
     - Based on:
        - Past likes
        - Friends activity
        - Content similarity

-  Recommendation Systems
   - Suggest:
     - People to follow
     - Posts to view

- Spam / Fake Account Detection
   - Identify bots and fake users

- Sentiment / Toxicity Detection
  - Detect abusive comments

- Viral Content Detection
   - Predict which content will explode

___

> “In the social media domain, the focus is on user interaction and network growth. We track events like likes, comments, shares, and follows to measure engagement and retention. These signals are used in real-time systems to rank feeds, recommend content, and detect spam. Key metrics include DAU/MAU, engagement rate, and retention.”
