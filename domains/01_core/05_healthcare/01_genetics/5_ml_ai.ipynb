#### Challenges

- High Dimensional Data
  - Millions of SNPs
  - Few samples

- Small Dataset Problem
   - Hard to collect genetic data

- Noise & Variability
   - Not all mutations matter
- Interpretability
   - Need biological meaning
- Ethical Issues
   - DNA is highly sensitive
