##### Typical Pipeline

```tree
Banking App / Payment Gateway
   ↓
Transaction Events
   ↓
Kafka (real-time ingestion)
   ↓
Spark / Streaming Engine
   ↓
Fraud Detection Model
   ↓
Decision (Approve / Block)
   ↓
Storage + Dashboard
```

____

- Key Requirement
  - Real-time processing is CRITICAL
    - You cannot detect fraud after 1 hour 
    - Must happen in milliseconds
