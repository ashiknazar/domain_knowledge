> “In the finance domain, especially in fraud detection, the focus is on identifying suspicious transactions in real time. Each transaction is captured as an event with attributes like amount, location, and device. We analyze patterns such as unusual spending or location changes using streaming systems and machine learning models to detect fraud and minimize financial risk.”

____

- “Why is recall important in fraud detection?”
   - “Because missing a fraudulent transaction leads to direct financial loss, so detecting as many fraud cases as possible is critical.”

- “What features help detect fraud?”
  - Amount deviation
  - Location change
  - Device change
  - Transaction frequency

- “What are challenges?”
  - Imbalanced data (fraud is rare)
  - Real-time processing constraints
  - False positives affecting user experience

____

## Problem: Imbalanced Data

### 1.Resampling Techniques

- Undersampling
   - Reduce normal transactions
      - Keep all fraud data
      - Remove some normal data


- Oversampling
   - Increase fraud data
      - Duplicate fraud samples
      - Or generate new ones

- SMOTE(Synthetic Minority Oversampling Technique)
   - Creates new fake fraud points
   - Based on existing fraud patterns

### 2. Use Better Evaluation Metrics

- Don’t use:
  - Accuracy
- Use:
  - Precision
  - Recall 
  - F1-score
  - ROC-AUC

- Key line:
  - “In fraud detection, recall is critical because missing a fraud case leads to financial loss.”

### 3. Cost-Sensitive Learning

- Not all mistakes are equal
- Missing fraud → VERY costly
- False alarm → less costly

### 4.Anomaly Detection (VERY POWERFUL)

- Instead of learning fraud…
- Learn normal behavior

##### Techniques:
- Isolation Forest
- Autoencoders
- One-Class SVM

> “Since fraud is rare, we can model normal behavior and detect deviations as anomalies.”

### 5.Ensemble Methods

- Combine multiple models
  - Random Forest
  - XGBoost

- These handle imbalance better with:
   - class weights
   - tree-based splits

### Real-time Rules + ML (Industry approach)

- Real systems don’t rely only on ML
- They combine:
- Rule-based:
  - Block if amount > threshold
  - Block if country changes suddenly
- ML-based:
  - Pattern detection

___

> “Fraud detection is an imbalanced problem where fraud cases are very rare. To handle this, we use techniques like oversampling (SMOTE) or undersampling to balance the data. We also rely on metrics like recall instead of accuracy. In addition, anomaly detection methods help identify unusual behavior, and in production systems, we combine rule-based and ML approaches for better performance.”
