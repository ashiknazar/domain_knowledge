#### Key KPIs

- Fraud Rate
  - % of transactions that are fraudulent

- Precision (Fraud detection)
   - Of predicted frauds, how many are actually fraud?

- Recall
  - Of actual frauds, how many did we catch?

- Transaction Success Rate
   - % of successful transactions

- Revenue / Fees
   - Charges from transactions

- Important Insight
  - Finance is risk-sensitive

#### Unlike e-commerce:
- Mistakes = money loss 💸
- False negatives (missing fraud) = very dangerous
