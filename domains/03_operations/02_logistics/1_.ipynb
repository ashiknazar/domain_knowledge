# 🚚 Logistics Domain Knowledge

## 🎯 Business Understanding

Logistics focuses on the **efficient movement of goods** from origin to destination while minimizing cost, reducing delays, and improving customer satisfaction.

### Core Goals

* Fast and reliable delivery
* Low transportation cost
* High fleet utilization
* Accurate shipment visibility
* Efficient last-mile delivery

---

## 📊 Key KPIs

### Delivery Performance

* **On-Time Delivery Rate (OTD)**
  `Delivered on time / Total deliveries`

* **Average Delivery Time**
  `Total delivery duration / Number of deliveries`

* **First Attempt Delivery Success Rate**
  `Successful first attempts / Total deliveries`

---

### Cost Metrics

* **Cost per Delivery**
  `Total logistics cost / Total deliveries`

* **Cost per Kilometer**
  `Transportation cost / Total distance traveled`

---

### Fleet Metrics

* **Fleet Utilization Rate**
  `Active vehicles / Total vehicles`

* **Capacity Utilization**
  `Used load capacity / Total capacity`

---

### Customer Metrics

* **Return Rate**
  `Returned shipments / Total shipments`

* **Customer Satisfaction Score**

---

## 🧠 Important Concepts

### 1. Last-Mile Delivery

Final delivery step from local hub to customer. Usually the most expensive and complex stage.

### 2. Lead Time

Total time from shipment creation to final delivery.

### 3. ETA (Estimated Time of Arrival)

Predicted delivery completion time.

### 4. Reverse Logistics

Handling returns, exchanges, damaged goods.

### 5. Cross Docking

Direct transfer from inbound truck to outbound truck with minimal storage.

### 6. Fleet Management

Managing vehicles, drivers, fuel, maintenance, schedules.

### 7. Cold Chain Logistics

Temperature-controlled transport for food, pharma, medical products.

---

## 📐 Important Formulas

### Reorder Point

Reorder\ Point = Demand\ During\ Lead\ Time + Safety\ Stock

### Fleet Utilization

Fleet\ Utilization = \frac{Active\ Vehicles}{Total\ Vehicles}

### Capacity Utilization

Capacity\ Utilization = \frac{Used\ Capacity}{Total\ Capacity}

### Cost per Delivery

Cost\ per\ Delivery = \frac{Total\ Logistics\ Cost}{Total\ Deliveries}

### On-Time Delivery Rate

OTD = \frac{On\ Time\ Deliveries}{Total\ Deliveries}
