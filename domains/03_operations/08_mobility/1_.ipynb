### Domain: Mobility

- Mobility means the movement of people or goods from one place to another, usually optimized through technology, data, and operations.

#### It includes:

- Ride-hailing
- Public transport
- Delivery fleets
- Logistics routing
- Vehicle sharing
- Traffic systems
- EV charging networks
- Navigation platforms

- **ride hailing** refers to on-demand transportation services where customers book rides through a digital platform (usually a mobile app) and are matched with nearby drivers using their own vehicles or fleet vehicles.

___

#### Why Mobility is Important in Data Careers

Mobility is a data-heavy domain because it constantly generates:

- GPS coordinates
- Trip requests
- Routes
- Traffic conditions
- Driver status
- Vehicle telemetry
- Demand patterns
- ETA estimates
- Payment events



___

### Main Subdomains of Mobility

### 1.Ride-Hailing
- Apps connecting riders and drivers.
- Examples:
  - Uber 
  - Ola
  - Lyft
- Problems solved:
  - Driver matching
  - ETA prediction
  - Surge pricing
  - Fraud detection


### 2.Public Transportation
- Bus, metro, rail, ferry systems.
- Problems solved:
  - Delay prediction
  - Passenger load forecasting
  - Route optimization

### 3.Last-Mile Delivery

- Food delivery / parcel delivery.
- Examples:
  - Swiggy
  - Zomato
  - Amazon delivery
- Problems solved:
  - Delivery time prediction
  - Rider allocation
  - Batch routing

### 4.Shared Mobility

- Bike rentals, scooters, car-sharing.
- Examples:
  - Lime
  - Bounce

- Problems solved:
  - Asset placement
  - Battery optimization
  - Theft risk

### 5.Fleet Management

- For logistics companies with many vehicles.
- Problems solved:
   - Fuel efficiency
   - Driver behavior
   - Vehicle maintenance
