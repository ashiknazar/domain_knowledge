### Common Data Sources

- **App Events**
  - app_open
  - ride_requested
  - ride_accepted
  - ride_started
  - ride_completed
  - payment_done

- **GPS / Location**
  - live coordinates
  - route traces
  - pickup/drop points
- **Operational Data**
  - driver online/offline
  - vehicle status
  - fuel/battery
- **External Data**
  - weather
  - traffic
  - events
  - holidays
  

### ML / AI Use Cases in Mobility

#### ETA Prediction
- Predict arrival time using:
  - traffic
  - distance
  - time of day
  - road conditions
#### Demand Forecasting
- Predict future ride demand in zones.
- Used for:
  - driver positioning
  - surge planning
#### Dynamic Pricing
- Increase price when demand > supply.

#### Route Optimization
- Find fastest / cheapest route.

#### Fraud Detection
- Examples:
  - fake trips
  - GPS spoofing
  - promo abuse
#### Driver-Rider Matching
- Best driver chosen based on:
  - proximity
  - acceptance history
  - route efficiency


#### Data Engineering in Mobility

- Typical pipeline:
```bash
Mobile App → Kafka → Spark/Flink → Data Lake → Dashboard / ML Models
```
- Real-time systems are critical because delays hurt customer experience.

### If Asked in Interview: “Explain Mobility Domain”

**Use this:**

> Mobility domain focuses on transporting people or goods efficiently using digital platforms. It includes ride-hailing, delivery, public transport, and fleet systems. Core metrics include wait time, trip completion, ETA accuracy, utilization, and revenue. Data comes from app events, GPS streams, and operational systems. Common ML use cases include ETA prediction, demand forecasting, dynamic pricing, and route optimization.
