### Key KPIs in Mobility

- Customer Side
  - Ride requests
  - Booking success rate
  - Wait time
  - ETA accuracy(Estimated Time of Arrival)
  - Cancellation rate
  - Customer rating

- Driver / Fleet Side
  - Driver utilization %
  - Trips per hour
  - Idle time
  - Earnings per day
  - Vehicle uptime

- Business Side
  - Revenue per trip
  - Cost per km
  - Retention rate
  - Region growth
