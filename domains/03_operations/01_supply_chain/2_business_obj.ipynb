- Goals
  - Deliver products on time
  - Reduce cost
  - Avoid stockouts
  - Optimize inventory

#### Core Trade-off
-  VERY IMPORTANT (interview gold)
   - Too much inventory → high cost
   - Too little inventory → lost sales

___

#### Core KPIs

- Inventory Level
  - How much stock is available

- Stockout Rate
   - % of time product is unavailable

- Inventory Turnover
   - How fast inventory is sold

- Lead Time
   - Time from order → deliver

- Fill Rate
   - % of orders fulfilled completely

- Logistics Cost
   - Cost of transportation + storage
