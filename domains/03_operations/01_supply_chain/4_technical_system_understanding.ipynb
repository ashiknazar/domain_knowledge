### Pipeline

```tree
ERP / Order System
   ↓
Event Stream (Kafka)
   ↓
Processing (Spark)
   ↓
Use Cases:
   - Demand forecasting
   - Inventory optimization
   - Delay detection
   ↓
Storage (Warehouse / HDFS)
   ↓
Dashboards / Alerts

___

#### Systems Involved
- ERP (Enterprise Resource Planning) Systems
   - Manage orders, inventory
- WMS (Warehouse Management System)
- TMS (Transport Management System)


___
