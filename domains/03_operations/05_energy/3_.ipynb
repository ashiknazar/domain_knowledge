### Major Data Sources in Energy

___

#### 1.Smart Meter Data

- Collected every 15 mins / hourly.
- Fields:
  - meter_id
  - timestamp
  - kWh usage
  - voltage
  - outage flag

### 2.SCADA Data

Supervisory Control and Data Acquisition systems.

Industrial real-time sensor systems.

- Contains:
  - pressure
  - temperature
  - load
  - switch status

#### 3. Grid Sensor Data
  - transformer load
  - line current
  - frequency
  - faults

### 4.Billing Systems
- bill amount
- due date
- payment status
- tariff type

#### 5.Weather Data

Critical for solar/wind and demand.
 - temperature
 - cloud cover
 - wind speed
 - humidity

### 6.Asset Maintenance Logs
- equipment type
- failure history
- service dates

## Core Concepts You Must Know

### Base Load vs Peak Load
  - Base Load
    - Minimum continuous demand.
  - Peak Load
    - Highest demand periods (evening often).

### Demand Response
  - Customers reduce usage during peak times.

>Demand Response = Consumers help balance the grid by using less power when demand is highest.

### Tariff
- Price structure for electricity.
- Examples:
  - residential
  - industrial
  - time-of-use pricing

### Net Metering

- Solar users sell excess energy back to grid.

#### Load Shedding
- Controlled power cuts when supply insufficient.

## ML / AI Use Cases in Energy

### Demand Forecasting

- Predict electricity demand by hour/day/city.
- Huge use case.
- Inputs:
  - weather
  - season
  - holidays
  - historical usage

### Predictive Maintenance
- Predict transformer / turbine failures before breakdown.

### Renewable Output Forecasting
- Solar / wind generation prediction.
- Depends on weather.

### Energy Theft Detection
- Illegal tapping / meter tampering.
- Patterns:
  - unusual drops
  - impossible usage curves

### Dynamic Pricing
- Higher prices at peak demand.

## EV Charger Optimization
