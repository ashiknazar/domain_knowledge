### Important KPIs in Energy

___

#### 1.Operational KPIs

- Capacity Utilization
   - How much plant capability is used.
- Plant Load Factor (PLF)
   - Actual generation / maximum possible generation.
   - Very common in power generation.

- Downtime
   - Hours system unavailable.

- Mean Time to Repair (MTTR)
  - Repair speed after failure.

___

#### 2.Grid KPIs
- SAIDI (System Average Interruption Duration Index)
  - Average outage duration per customer.
- SAIFI
  - Average outage frequency.
- Transmission Loss %
  - Energy lost during movement.
- Voltage Stability
  - Power quality consistency.


- **SAIDI** = System Average Interruption Duration Index  
- **SAIFI** = System Average Interruption Frequency Index

### Grid KPI Formulas

- $SAIDI = \frac{\text{Total Customer Interruption Duration}}{\text{Total Number of Customers Served}}$

- $SAIFI = \frac{\text{Total Number of Customer Interruptions}}{\text{Total Number of Customers Served}}$

- $Transmission\ Loss\% = \frac{\text{Energy Sent} - \text{Energy Received}}{\text{Energy Sent}} \times 100$

- $Voltage\ Stability = \text{Ability to maintain acceptable voltage levels under normal and disturbed conditions}$


___

#### 3.Customer KPIs
- Billing Collection Rate
   - Paid bills / issued bills.
- Complaint Resolution Time
- Customer Satisfaction
- Churn Rate (retail energy markets)

#### 4.Commercial KPIs
- Cost per kWh
- Revenue per Customer
- Peak Demand Cost
- Margin by Tariff Plan

### 5.Sustainability KPIs

- Renewable Mix %
  - % power from renewables.
- Carbon Emission
- Energy Efficiency Score
