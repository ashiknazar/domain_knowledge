> The telecom domain is the industry that enables communication between people, devices, and systems through:

- Mobile networks (2G / 3G / 4G / 5G)
- Internet / broadband
- Voice calls
- SMS
- Fiber networks
- Enterprise connectivity
- IoT connectivity

- Examples:
  - Jio
  - Airtel
  - Vi
  - BSNL
  - Vodafone
  - AT&T
  - Verizon

### Why Telecom Is Important for Data Careers
- Telecom is one of the richest data domains because it generates huge-scale real-time data:

  - Billions of network events
  - Customer usage data
  - Recharge/payments
  - Tower performance logs
  - Call records
  - Device signals
  - Roaming data
  - Support interactions

### Core Telecom Business Segments

- **1.Consumer Mobile Services**
- What users buy:
  - Prepaid SIM
  - Postpaid plans
  - Data packs
  - Voice plans
  - Family plans

-  **2.Broadband / Fiber**
  - Home internet:
     - FTTH
     - Wi-Fi services
     - Business broadband

- **3.Enterprise Connectivity**
- Services for businesses:
  - Dedicated internet
  - VPN
  - Cloud connectivity
  - SIP voice systems

- **4.Infrastructure**
- Physical assets:
  - Towers
  - Fiber cables
  - Switches
  - Routers
  - Data centers

-  **5.IoT / M2M**
- SIM connectivity for:
  - Smart meters
  - Vehicles
  - POS machines
  - Cameras
