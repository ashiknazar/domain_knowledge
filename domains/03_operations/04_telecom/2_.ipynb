### Telecom KPIs (Very Important)

#### 1. Revenue KPIs

- ARPU = Average Revenue Per User
  - Revenue / Active Users

- Recharge Frequency
  - How often prepaid users recharge.
- Plan Upgrade Rate
  - How many users move to higher-value plans.

### Customer KPIs
- Churn Rate
  - Customers leaving network.
- Retention Rate
  - Customers staying active.
- NPS / Satisfaction
  - Net promoter score.
- Complaint Rate
  - Support complaints per 1000 users.

### Network KPIs

- Call Drop Rate
  - Calls disconnected unexpectedly.
- Network Availability
  - Uptime of towers / systems.
- Latency
  - Delay in network response.
- Throughput
  - Download/upload speed.
- Coverage %
  - Geographic reach.

### Usage KPIs
- Data Consumption per User
  - GB/month.
- Voice Minutes
  - Total talk time.
- SMS Volume
  - Legacy but still relevant in some markets.

### Telecom Data Sources
### CDR (Call Detail Records)

- Extremely important.
- Contains:
  - caller number
  - receiver number
  - start time
  - duration
  - tower ID
  - call status
  - roaming flag

- Used for analytics, fraud, billing.

### Data Session Logs
- Internet usage sessions:
  - bytes used
  - duration
  - app category
  - device type

### Recharge / Billing Data
- amount
- date
- channel
- success/failure
- autopay

### CRM Data
- customer demographics
- complaints
- support tickets
- plan history
### Network Logs
- tower alarms
- congestion
- packet loss
- downtime

### Device Data
- handset model
- OS version
- 4G/5G capable
- SIM type

### important Telecom Concepts
- Prepaid vs Postpaid

- Prepaid
  - Pay first, use later.
  - India heavily prepaid historically.

- Postpaid
  - Use first, billed monthly.
